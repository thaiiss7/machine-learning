```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
base_estudantes = pd.read_csv("student_exam_scores.csv")
base_estudantes
```

```python
base_estudantes.isnull().sum()
```

```python
base_estudantes.describe()
```

```python
base_estudantes.drop('student_id', axis= 1, inplace=True)
```

```python
base_estudantes
```

```python
figure = plt.figure(figsize=(5,5))
sns.heatmap(base_estudantes.corr(), annot=True)
```

```python
colunas = ['hours_studied', 'sleep_hours', 'attendance_percent', 'previous_scores', 'exam_score']

plt.figure(figsize=(10,10))

for i, coluna in enumerate(colunas, 1):
    plt.subplot(len(colunas), 1, i)
    sns.boxplot(data=base_estudantes, x=coluna, fliersize=3, linewidth=1)
    plt.title(coluna, fontsize=11)
    plt.xlabel('')
    plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
```

```python
plt.hist(base_estudantes["sleep_hours"])
plt.xlabel("sleep_hours")
plt.ylabel("exam_score")
```

```python
x_estudantes = base_estudantes.iloc[:, :4].values
y_estudantes = base_estudantes.iloc[:, 4].values
```

```python
x_estudantes
```

```python
y_estudantes
```

```python
from sklearn.model_selection import train_test_split
x_estudantes_treinamento, x_estudantes_teste, y_estudantes_treinamento, y_estudantes_teste = train_test_split(x_estudantes, y_estudantes, test_size=0.2, random_state=0)
```

```python
x_estudantes_treinamento.shape, x_estudantes_teste.shape
```

#Regressão Linear Múltipla

```python
from sklearn.linear_model import LinearRegression
regressor_multiplo_estudantes = LinearRegression()
regressor_multiplo_estudantes.fit(x_estudantes_treinamento, y_estudantes_treinamento)
```

```python
regressor_multiplo_estudantes.score(x_estudantes_treinamento, y_estudantes_treinamento)
```

```python
regressor_multiplo_estudantes.score(x_estudantes_teste, y_estudantes_teste)
```

```python
previsoes = regressor_multiplo_estudantes.predict(x_estudantes_teste)
previsoes
```

```python
y_estudantes_teste
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_estudantes_teste, previsoes)
mae
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mse = mean_squared_error(y_estudantes_teste, previsoes)
mse
```

#Regressão Polinomial

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree= 2)
```

```python
x_estudantes_treinamento_poly = poly.fit_transform(x_estudantes_treinamento)
x_estudantes_teste_poly = poly.transform(x_estudantes_teste)
```

```python
regressor_estudantes_poly = LinearRegression()
regressor_estudantes_poly.fit(x_estudantes_treinamento_poly, y_estudantes_treinamento)
```

```python
regressor_estudantes_poly.score(x_estudantes_treinamento_poly, y_estudantes_treinamento)
```

```python
regressor_estudantes_poly.score(x_estudantes_teste_poly, y_estudantes_teste)
```

```python
previsoes = regressor_estudantes_poly.predict(x_estudantes_teste_poly)
previsoes
```

```python
y_estudantes_teste
```

```python
mean_absolute_error(y_estudantes_teste, previsoes)
```

```python
mean_squared_error(y_estudantes_teste, previsoes)
```

```python
regressor_arvore_estudantes.score(x_estudantes_treinamento, y_estudantes_treinamento)
```
