```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier, plot_tree
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = load_breast_cancer()
description = data.DESCR.split('\n')
print('\n'.join(description[11:31]))
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.shape
```

```python
sns.countplot(x= 'target', data= df, color= 'maroon')
plt.title('Target Variable Distribution')
plt.xlabel('0: Benigno, 1: Maligno')
plt.ylabel('Score')
plt.show()
```

```python
clf = DecisionTreeClassifier(max_depth=2)
x = data.data
y = data.target
clf.fit(x,y)
plt.figure(figsize=(10,8))
plot_tree(clf, filled=True, feature_names=data.feature_names, class_names=data.target_names)
plt.show()
```

```python
features = df.columns[:-1]
fig, axs = plt.subplots(nrows=5, ncols=6, figsize=(20,15))
axs = axs.flatten()
for i, feature in enumerate(features):
    sns.violinplot(x='target', y=feature, data=df, split=False, ax=axs[i])
    axs[i].set_title(f"Distribution of {feature}")
    axs[i].set_xlabel('0: Benign, 1:Malignant')
    axs[i].set_ylabel(feature)
    
for i in range(len(features), len(axs)):
    axs[i].remove()
    
plt.tight_layout()
plt.show()
```

```python
x = df.drop(columns=['target'])
y = df['target']
x
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x = scaler.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', C= 2)
svm.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

previsoes = svm.predict(x_test)
accuracy_score(y_test, previsoes)
```

PCA - principal components analysis

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=5)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)
x_train_pca
```

```python
svm_pca = SVC(kernel='linear', C= 2)
svm_pca.fit(x_train_pca, y_train)
```

```python
previsoes = svm_pca.predict(x_test_pca)
previsoes
```

```python
accuracy_score(y_test, previsoes)
```

```python
from sklearn.feature_selection import f_classif, SelectKBest

selection = SelectKBest(f_classif, k=7)
x_anova = selection.fit_transform(x,y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_anova, y, test_size=0.3, random_state=42)
```

```python
svm_anova = SVC(kernel='linear', C=2)
svm_anova.fit(x_train, y_train)
```

```python
previsoes = svm_anova.predict(x_test)
accuracy_score(y_test, previsoes)
```

```python
from sklearn.model_selection import GridSearchCV
parametros = {'C': [1.0, 1.5, 2.0],
            'kernel': ['rbf', 'linear', 'poly', 'sigmoid']}
```

```python
grid_search = GridSearchCV(estimator=SVC(), param_grid=parametros)
grid_search.fit(x,y)
best_parames = grid_search.best_params_
best_result = grid_search.best_score_
print(f"melhores paramestros; {best_parames}")
print(f"melhor resultado: {best_result}")
```

```python
parametros = {'n_estimators': [10, 40, 100, 150, 200]}
```

```python
from sklearn.ensemble import RandomForestClassifier
grid_search = GridSearchCV(estimator=RandomForestClassifier(), param_grid=parametros)
grid_search.fit(x,y)
best_parames = grid_search.best_params_
best_result = grid_search.best_score_
print(f"melhores paramestros: {best_parames}")
print(f"melhor resultado: {best_result}")
```

```python
from sklearn.model_selection import cross_val_score, KFold

results_svm = []

for i in range(30):
    kfold = KFold(n_splits=10, shuffle=True, random_state=i)
    svm = SVC(kernel= 'rbf', C=1.5)
    scores = cross_val_score(svm, x , y, cv=kfold)
    results_svm.append(scores.mean())
```
